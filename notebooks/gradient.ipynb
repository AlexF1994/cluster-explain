```python
import matplotlib.pyplot as plt
from sklearn import datasets
import mpl_toolkits.mplot3d 
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import pandas as pd
from cxplain.xkm import XkmExplainer
from cxplain.gradient import GradientExplainer  
from cxplain.metrics import EuclideanMetric, Metric, ManhattenMetric
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
kmeans = KMeans(n_clusters=3, random_state=3).fit(X)
cluster_centers = kmeans.cluster_centers_
```

```python
cluster_centers
```

```python
predictions = kmeans.predict(X)
predictions
```

```python
np.take(cluster_centers, predictions, axis=0)
```

```python
explainer = GradientExplainer(X, cluster_centers, predictions, ManhattenMetric, enable_abs_calculation=False)
```

```python
exp = explainer.fit_explain()
```
